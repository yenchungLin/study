```python
import pandas as pd 
import xml.etree.ElementTree as et     
xtree = et.parse("/Users/yulin/Desktop/高公局/cctv_info_0000.xml")
xroot = xtree.getroot() 

cctvid = []
roadsection = []
locationpath = []
startlocationpoint = []
endlocationpoint = []
longitudes = []
latitudes = []

for i in xroot.findall("Infos"):
    for j in i.findall("Info"):
        cctvid.append(j.get("cctvid"))
        roadsection.append(j.get("roadsection"))
        locationpath.append(j.get("locationpath"))
        startlocationpoint.append(j.get("startlocationpoint"))
        endlocationpoint.append(j.get("endlocationpoint"))
        longitudes.append(j.get("px"))
        latitudes.append(j.get("py"))

df = pd.DataFrame()
df["cctvid"] = cctvid
df["roadsection"] = roadsection
df["locationpath"] = locationpath
df["startlocationpoint"] = startlocationpoint
df["endlocationpoint"] = endlocationpoint
df["longitude"] = longitudes
df["latitude"] = latitudes
df
```
