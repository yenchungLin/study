```python
Folder_Path = r'/Users/yulin/Documents/M05A'          #要拼接的文件夹及其完整路径，注意不要包含中文
SaveFile_Path =  r'/Users/yulin/Documents/M05A'       #拼接后要保存的文件路径
SaveFile_Name = r'M05A2016.csv'
#修改当前工作目录
os.chdir(Folder_Path)
#将该文件夹下的所有文件名存入一个列表
file_list = os.listdir()
for i in range(len(file_list)):
    Folder_Path_1 = Folder_Path +'/'+ file_list[i]       #要拼接的文件夹及其完整路径，注意不要包含中文
    os.chdir(Folder_Path_1)
    file_list_1 = os.listdir()
    for j in range(len(file_list_1)):
        Folder_Path_2 = Folder_Path_1 +'/'+ file_list_1[j]       #要拼接的文件夹及其完整路径，注意不要包含中文
        os.chdir(Folder_Path_2)
        file_list_2 = os.listdir()
        df0 = pd.read_csv(Folder_Path_2 + '/'+ file_list_2[0], header=None,names=['A', 'B','C','D','E','F'])
        for k in range(1,6):
            df1 = pd.read_csv(Folder_Path_2 + '/'+ file_list_2[k],header=None,names=['G', 'B','C','D','H','I'])
            df0 = pd.merge(df0, df1,left_on=['B','C','D'] ,right_on=['B','C',"D"],how='inner')
            df0["E"] = df0[["E","H"]].sum(1)
            df0["F"] = df0[["F","I"]].sum(1)
            df0 = df0.drop(columns=['G', 'H','I'])
        df0["E"] = df0["E"] / 6
        df0["G"] = df0["F"] / df0["E"]
        df0 = df0.fillna(0)
        df0.to_csv(SaveFile_Path+'\\'+ SaveFile_Name,encoding="utf_8_sig", header=None,index=False, mode='a+')
        df0 = pd.read_csv(Folder_Path_2 + '/'+ file_list_2[6], header=None,names=['A', 'B','C','D','E','F'])
        for l in range(7,12):
            df1 = pd.read_csv(Folder_Path_2 + '/'+ file_list_2[l],header=None,names=['G', 'B','C','D','H','I'])
            df0 = pd.merge(df0, df1,left_on=['B','C','D'] ,right_on=['B','C',"D"],how='inner')
            df0["E"] = df0[["E","H"]].sum(1)
            df0["F"] = df0[["F","I"]].sum(1)
            df0 = df0.drop(columns=['G', 'H','I'])
        df0["E"] = df0["E"] / 6
        df0["G"] = df0["F"] / df0["E"]
        df0 = df0.fillna(0)
        df0.to_csv(SaveFile_Path+'\\'+ SaveFile_Name,encoding="utf_8_sig", header=None,index=False, mode='a+')
```
