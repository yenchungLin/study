```python
class tree:
    def __init__(self):
        self.data = 0
        self.left = None
        self.right = None

def create_tree(root,val):
    print('root = ',root,'val=',val)
    newnode = tree()
    newnode.data = val
    newnode.left = None
    newnode.right = None
    print('newnode.data = ',newnode.data)
    current=tree()
    if root == None:
        root = newnode
        print('root.data = ',root.data,'root.left = ',root.left,'root.right = ',root.right)
        return root
        
    else:
        print('before_while_root.data = ',root.data)
        current = root
        
        while current != None:
            #print('一開始current.data = ',current.data)
            backup = current
            print('一開始current.data = ',current.data)
            #節點的值大小比較，決定放在左邊或是右邊
            if current.data > val:
                #print('current.left = ',current.left)
                current = current.left
                print('while_current.left = ',current.left)
            else:
                #print('current.right = ',current.right)
                current = current.right
                print('while_current.right = ',current.right)
            print('執行中current.data = ',current.data)
        
        if backup.data > val:
            backup.left = newnode
        else:
            backup.right = newnode
    return root
```

```python
data = [5,6,4,8,12,3,17,1,9]
#data = [0,6,3,5,4,7,8,9,2]
ptr = None
root = None
for i in range(9):
    if i ==0:
        print('i=',i,'data[i]= ',data[i],'ptr = ',ptr,'root = ',root)
    else:
        print('i=',i,'data[i]= ',data[i],'ptr.data = ',ptr.data,'root = ',root)
    ptr = create_tree(ptr,data[i])
    print('ptr.data = ',ptr.data,'ptr.left = ',ptr.left,'ptr.right = ',ptr.right)
print('左子樹：')
root = ptr.left
while root != None:
    print('%d'%root.data)
    root = root.left
print('-------------------------')
print('右子樹：')
root = ptr.right
while root != None:
    print('%d'%root.data)
    root = root.right
print()

for i in range(9):
    print('data[i].left = ',ptr[i])
```
